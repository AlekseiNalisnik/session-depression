<h1>Обработка данных До сессии</h1>

```python
import pandas as pd

data_before = pd.read_csv('sample_before_session.csv')

df_before = pd.DataFrame(data_before)
df_before.columns.values
```

```python
anxiety_before = []
for index, row in df_before.iterrows():
    score = int(row['1. Я испытываю напряжение, мне не по себе'][0]) \
    + int(row['2. Я испытываю страх, кажется, что-то ужасное может вот-вот случиться.'][0]) \
    + int(row['3. Беспокойные мысли крутятся у меня в голове'][0]) \
    + int(row['4. Я легко могу расслабиться'][0]) \
    + int(row['5. Я испытываю внутреннее напряжение или дрожь'][0]) \
    + int(row['6. Я испытываю неусидчивость, мне постоянно нужно двигаться'][0]) \
    + int(row['7. У меня бывает внезапное чувство паники'][0])
    anxiety_before.append(score)
```

<h1>Часть 1 (До и после сессии)</h1>
<b>Задача:</b> необходимо проверить гипотезу о том, что средние значения двух генеральных совокупностей,
из которых извлечены сравниваемые независимые выборки, отличаются друг от друга.
<br>
<i>Допущение независимости предполагает, что представители двух выборок не составляют пары коррелирующих значений признака.</i>
<br>
Проверяемая статистическая гипотеза $Н_{0}$: $\bar M_1$ = $\bar M_2$.
<br>
При ее отклонении принимается альтернативная гипотеза о том, что $\bar M_1$ $\neq$ $\bar M_2$.
<br>
<i>Для решения будем использовать <b>t-Критерий Стьюдента для независимых выборок</b></i>
<hr />
Уровень значимости $\alpha$=0.01
<br />
$\bar M$ - выборочное среднее

```python
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Будем считать выбросами все точки, выходящие за пределы трёх сигм.
def remove_outliers(parameter, df):
    df_mean = df[parameter].mean()
    df_std = df[parameter].std()
    wrong = df[(df[parameter] < df_mean - 3*df_std) | (df[parameter] > df_mean + 3*df_std)].index
    out = set(wrong)
    print('Количество выбросов в выборке ({0}): {1}'.format(parameter, len(out)))
    df.drop(out, inplace=True)
    return df_mean, df_std

def calculateStatistics(parameter, df_first, df_second, label_first, label_second):
    df_first_copy = df_first.copy(deep=True)
    df_second_copy = df_second.copy(deep=True)

    df_first_mean, df_first_std = remove_outliers(parameter, df_first_copy)
    df_second_mean, df_second_std = remove_outliers(parameter, df_second_copy)

    print('\n{0}: ({1})'.format(label_first, parameter))
    print('Выборочное среднее: ', df_first_mean)
    print('Среднеквадратичное отклонение: ', df_first_std)
    print('\n{0}: ({1})'.format(label_second, parameter))
    print('Выборочное среднее: ', df_second_mean)
    print('Среднеквадратичное отклонение: ', df_second_std)

    fig, ax = plt.subplots(1, 2, sharex='col', sharey='row', figsize=(8, 5))
    df_first_copy[parameter].hist(bins=20, ax=ax[0])
    df_second_copy[parameter].hist(bins=20, ax=ax[1])

    # По центральной предельной теореме будем считать, что у нас распределение выборочного среднего асимптотически нормально
    statistic, pValue = ttest_ind(df_first_copy[parameter], df_second_copy[parameter])
    print('\nT-статистика: ', statistic)
    print('Уровень значимости: ', pValue)
    print()
```

<h2>1.1. Разница у всех-всех до и после сесии</h2>

```python
calculateStatistics('Тревожность', df_before, df_after, 'До сессии', 'После сессии')
```

```python
calculateStatistics('Депрессия', df_before, df_after, 'До сессии', 'После сессии')
```

<b>Результат (тревожность):</b> исходя из полученных данных можно сделать вывод о том, что статистическая гипотеза о равенстве выборочных средних значений тревожности отклоняется, так как уровень значимости меньше 0.01, поэтому с вероятностью 99% гипотеза $Н_{0}$: $\bar M_1$ = $\bar M_2$ отвергается, а принимается альтернативная ей гипотеза, в частности можно сделать вывод о том, что <i>уровень тревожности "до сессии" выше уровня тревожности "после сессии"</i>
<br />
<b>Результат (депрессия):</b> исходя из полученных данных можно сделать вывод о том, что статистическая гипотеза о равенстве выборочных средних значений депрессии также отклоняется, как и в случае с <i>тревожностью</i>, так как уровень значимости меньше 0.01, поэтому с вероятностью 99% гипотеза $Н_{0}$: $\bar M_1$ = $\bar M_2$ отвергается, а принимается альтернативная ей гипотеза, в частности можно сделать вывод о том, что <i>уровень депрессии "до сессии" выше уровня депрессии "после сессии"</i>

<h2>3.2. По отношению к долгам среди несдавших</h2>
<b>Для корректности и возможности расчётов все ответы в произвольной форме были классифицированы
на две группы: <i><q>Трепетно</q> и <q>Нейтрально</q></i></b>

Для оценки различий средних значений 2х независимых выборок в данном случае нам следует использовать для расчётов
<i>U-критерий Манна-Уитни</i>, так как размеры каждой из выборок не превышают 25 элементов, и считать распределение
асимптотически нормальным невозможно

```python
from scipy.stats import mannwhitneyu

# Учтём, что данный критерий не особо чувствителен к выбросам
def calculateMannWhitneyUStatistics(parameter, df_first, df_second, label_first, label_second):
    df_first_mean = df_first[parameter].mean()
    df_first_std = df_first[parameter].std()
    df_second_mean = df_second[parameter].mean()
    df_second_std = df_second[parameter].std()

    print('\n{0}: ({1})'.format(label_first, parameter))
    print('Выборочное среднее: ', df_first_mean)
    print('Среднеквадратичное отклонение: ', df_first_std)
    print('\n{0}: ({1})'.format(label_second, parameter))
    print('Выборочное среднее: ', df_second_mean)
    print('Среднеквадратичное отклонение: ', df_second_std)

    fig, ax = plt.subplots(1, 2, sharex='col', sharey='row', figsize=(8, 5))
    df_first[parameter].hist(bins=20, ax=ax[0])
    df_second[parameter].hist(bins=20, ax=ax[1])

    # По центральной предельной теореме будем считать, что у нас распределение выборочного среднего асимптотически нормально
    statistic, pValue = mannwhitneyu(df_first[parameter], df_second[parameter])
    print('\nU-статистика Манна-Уитни: ', statistic)
    print('Уровень значимости: ', pValue)
    print()
```

```python
calculateMannWhitneyUStatistics(
    'Депрессия',
    df_after.loc[df_after['Отношение к долгам среди несдавших'] == 'Трепетно'],
    df_after.loc[df_after['Отношение к долгам среди несдавших'] == 'Нейтрально'],
    'Трепетно',
    'Нейтрально'
)
```

```python
calculateMannWhitneyUStatistics(
    'Тревожность',
    df_after.loc[df_after['Отношение к долгам среди несдавших'] == 'Трепетно'],
    df_after.loc[df_after['Отношение к долгам среди несдавших'] == 'Нейтрально'],
    'Трепетно',
    'Нейтрально'
)
```
