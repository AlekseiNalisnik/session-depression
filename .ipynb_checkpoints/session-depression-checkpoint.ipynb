<h3>Часть 1</h3>
<b>Задача:</b> необходимо проверить гипотезу о том, что средние значения двух генеральных совокупностей,
из которых извлечены сравниваемые независимые выборки, отличаются друг от друга.
<br>
<i>Допущение независимости предполагает, что представители двух выборок не составляют пары коррелирующих значений признака.</i>
<br>
Проверяемая статистическая гипотеза $Н_{0}$: $\bar M_1$ = $\bar M_2$.
<br>
При ее отклонении принимается альтернативная гипотеза о том, что $\bar M_1$ $\neq$ $\bar M_2$.
<br>
<i>Для решения будем использовать <b>t-Критерий Стьюдента для независимых выборок</b></i>

```python
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Будем считать выбросами все точки, выходящие за пределы трёх сигм.
def remove_outliers(parameter, df):
    df_mean = df[parameter].mean()
    df_std = df[parameter].std()
    wrong = df[(df[parameter] < df_mean - 3*df_std) | (df[parameter] > df_mean + 3*df_std)].index
    out = set(wrong)
    print('Количество выбросов в выборке ({0}): {1}'.format(parameter, len(out)))
    df.drop(out, inplace=True)
    return df_mean, df_std

def calculateStatistics(parameter):
    df_before_copy = df_before.copy(deep=True)
    df_after_copy = df_after.copy(deep=True)

    df_before_mean, df_before_std = remove_outliers(parameter, df_before_copy)
    df_after_mean, df_after_std = remove_outliers(parameter, df_after_copy)

    print('\nДо сессии ({0})'.format(parameter))
    print('Выборочное среднее: ', df_before_mean)
    print('Среднеквадратичное отклонение: ', df_before_std)
    print('\nПосле сессии ({0})'.format(parameter))
    print('Выборочное среднее: ', df_after_mean)
    print('Среднеквадратичное отклонение: ', df_after_std)

    fig, ax = plt.subplots(1, 2, sharex='col', sharey='row', figsize=(8, 5))
    df_before_copy[parameter].hist(bins=20, ax=ax[0])
    df_after_copy[parameter].hist(bins=20, ax=ax[1])

    # По центральной предельной теореме будем считать, что у нас распределение выборочного среднего асимптотически нормально
    statistic, pValue = ttest_ind(df_before_copy[parameter], df_after_copy[parameter])
    print('\nT-статистика: ', statistic)
    print('Уровень значимости: ', pValue)
    print()
```

```python
calculateStatistics('Тревожность')
```

<b>Результат:</b> исходя из полученных данных можно сделать вывод о том, что статистическая гипотеза о равенстве выборочных средних значений тревожности отклоняется, так как уровень значимости меньше 0.01, поэтому с вероятностью 99.99% гипотеза $Н_{0}$: $\bar M_1$ = $\bar M_2$ отвергается, а принимается альтернативная ей гипотеза, в частности можно сделать вывод о том, что <i>уровень тревожности "до сессии" выше уровня тревожности "после сессии"</i>

```python
calculateStatistics('Депрессия')
```

<b>Результат:</b> исходя из полученных данных можно сделать вывод о том, что статистическая гипотеза о равенстве выборочных средних значений депрессии также отклоняется, как и в случае с <i>тревожностью</i>, так как уровень значимости меньше 0.01, поэтому с вероятностью 99.99% гипотеза $Н_{0}$: $\bar M_1$ = $\bar M_2$ отвергается, а принимается альтернативная ей гипотеза, в частности можно сделать вывод о том, что <i>уровень депрессии "до сессии" выше уровня депрессии "после сессии"</i>
