```python
import pandas as pd
```

```python
data = pd.read_csv('sample_before_session.csv')
```

```python
df = pd.DataFrame(data)
df.columns.values
```

```python
anxiety = []
for index, row in df.iterrows():
    score = int(row['1. Я испытываю напряжение, мне не по себе'][0]) \
    + int(row['2. Я испытываю страх, кажется, что-то ужасное может вот-вот случиться.'][0]) \
    + int(row['3. Беспокойные мысли крутятся у меня в голове'][0]) \
    + int(row['4. Я легко могу расслабиться'][0]) \
    + int(row['5. Я испытываю внутреннее напряжение или дрожь'][0]) \
    + int(row['6. Я испытываю неусидчивость, мне постоянно нужно двигаться'][0]) \
    + int(row['7. У меня бывает внезапное чувство паники'][0])
    anxiety.append(score)
```

```python
depression = []
for index, row in df.iterrows():
    score = int(row['1. То, что раньше приносило мне большое удовольствие, и сейчас вызывает у меня такое же чувство'][0]) \
    + int(row['2. Я способен рассмеяться и увидеть в том или ином смешное'][0]) \
    + int(row['3. Я испытываю бодрость'][0]) \
    + int(row['4. Мне кажется, что я стал все делать очень медленно'][0]) \
    + int(row['5. Я не слежу за своей внешностью '][0]) \
    + int(row['6. Я считаю, что мои дела (занятия, увлечения) могут принести мне чувство удовлетворения'][0]) \
    + int(row['7. Я могу получить удовольствие от хорошей книги, радио- или телепрограммы'][0])
    depression.append(score)
```

```python
df['Тревожность'] = anxiety
df['Депрессия'] = depression
df
```

```python
df.to_excel("Результаты.xlsx")
```
